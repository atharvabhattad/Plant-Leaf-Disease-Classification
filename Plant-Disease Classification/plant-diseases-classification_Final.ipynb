```python
import tensorflow as tf                                
import matplotlib.pyplot as plt                                         
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam         
from tensorflow.keras.callbacks import EarlyStopping   
from tensorflow.keras.regularizers import l1, l2      
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, Flatten, Dropout,BatchNormalization  
from tensorflow.keras.applications import DenseNet121
```

```python
train_data = tf.keras.utils.image_dataset_from_directory(
    "/kaggle/input/new-plant-diseases-dataset/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/train/",
    labels='inferred',
    label_mode='categorical',
    image_size=(256, 256),
    batch_size=32)

train_data = train_data.map(lambda x, y: (x / 255.0, y))
```

```python
import os

# Path to the dataset directory
dataset_path = '/kaggle/input/new-plant-diseases-dataset/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/train'

# Get the class names (subdirectory names)
class_names = sorted(os.listdir(dataset_path))

# Print the class names
for idx, class_name in enumerate(class_names):
    print(f"{class_name},")
```

```python
val_data = tf.keras.preprocessing.image_dataset_from_directory(
    "/kaggle/input/new-plant-diseases-dataset/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/valid",
    labels='inferred',
    label_mode='categorical',
    image_size=(256, 256),
    batch_size=32)

val_data = val_data.map(lambda x, y: (x / 255.0, y))
```

```python
conv = DenseNet121(
    weights='imagenet',
    include_top = False,
    input_shape=(256,256,3),
    pooling='avg'
)
```

```python
conv.trainable = False
```

```python
model = Sequential()
model.add(conv)
model.add(BatchNormalization())
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.3))
model.add(BatchNormalization())
model.add(Dense(128, activation='relu'))
model.add(Dense(38, activation='softmax'))
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(train_data, epochs=20, validation_data=val_data, callbacks=[EarlyStopping(patience=0)], verbose=2)
```

```python
evaluation = model.evaluate(val_data)
print("Validation Loss:", evaluation[0])
print("Validation Accuracy:", evaluation[1])
```
